```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import os
from pathlib import Path

#sql
import sqlite3
```

```python
#local diretorio
local = './dataset/ajustados_2022'
dir_path = Path(local)


type(dir_path)
```

```python
files_loc = dir_path.iterdir()
files_loc
```

```python
# Lista arquivos
files_loc = dir_path.iterdir()

#converte bytes para mega
def bytes_mega(bytes):
    megabytes = bytes / (1024*1024)
    return megabytes

# somando o tam do arquivos
soma_mega = 0

# loop arquivos
for i in files_loc:
    #captura o tamanho dos arquivos
    tam_file = os.path.getsize(i)
    
    ##converte bytes para mega
    converte = bytes_mega(tam_file)
    
    ##melhorando a visualização
    nome_file = str(i).split('/')[-1]
    #print(nome_file)
#
    soma_mega = soma_mega + tam_file
    print(nome_file, '→', round(converte, 2),'MB')
    print('-'*30)
    #print(round(converte, 2))

print(f'Tamanho total de arquivos: {round(bytes_mega(soma_mega))} MBs de dados')
```

```python
#Consolidando os arquivos

# Lista arquivos
files_loc = dir_path.iterdir()


for contador, arq in enumerate(files_loc):
    print(contador, arq)

    if contador == 0:
        #Ler o arquivos
        data_base = pd.read_csv(arq, sep=';', encoding = 'latin-1')
        #salvar na variavel
        data_consolidada = data_base
    
    else:
        #Ler o arquivos
        data_base = pd.read_csv(arq, sep=';', encoding = 'latin-1')
        #concatena os dados
        data_consolidada = pd.concat([data_consolidada, data_base])

data_consolidada.shape

```

```python
data_consolidada = data_consolidada.sort_values(by='Data da Infração (DD/MM/AAAA)', ascending=True)
```

```python
#criar conexao
conect = sqlite3.connect('banco_dados.db')

#Tabela vazia
nome_tabela = 'base_consolidada'
```

```python
# Enviando os dados para o banco de datos
data_consolidada.to_sql(nome_tabela, conect, if_exists='append', index=False)
```

```python
#Para corrigir o erro na compressao dos arquivos vou transf. as colunas em str
for col in data_consolidada.columns:
    data_consolidada[col] = data_consolidada[col].astype(str)

#É muito comum na ingestão de dados o usúario alterar os tipos dos dados durante o preenchimento das info.
```

```python
## Criando o arquivo parquet para diminuir o tamanho dos dados
data_consolidada.to_parquet('base_consolidada.parquet', index=False, compression='gzip')
```
